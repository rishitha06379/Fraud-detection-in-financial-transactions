```python
import pandas as pd
from table_evaluator import TableEvaluator


original_data = pd.read_csv(r"Desktop/Jupyter/MP/processed_fraud_data.csv")
synthetic_data = pd.read_csv(r"Desktop/Jupyter/MP/balanced_WGAN.csv")  


categorical_columns = ['Device_Type', 'Transaction_Type', 'Card_Type', 'Authentication_Method']


min_rows = min(len(original_data), len(synthetic_data))
original_data_sampled = original_data.sample(n=min_rows, random_state=42)
synthetic_data_sampled = synthetic_data.sample(n=min_rows, random_state=42)


table_evaluator = TableEvaluator(original_data_sampled, synthetic_data_sampled, cat_cols=categorical_columns)


table_evaluator.visual_evaluation()

table_evaluator.plot_cumulative_difference()


table_evaluator.plot_distribution_comparison()
```
